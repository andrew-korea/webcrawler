```python
import requests
from bs4 import BeautifulSoup

URL = 'https://m.daum.net/'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
titles = soup.find_all('span', class_='tit_news') #title_news
```

```python
print(type(titles))
```

```python
soup.select_one('.tit_news').get_text() #only get text
```

```python
for i, title in enumerate(titles):
    print(i,title.get_text())
```
